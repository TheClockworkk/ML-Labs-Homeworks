1. Метрики качества классификации в зависимости от количества деревьев

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=5, n_informative=2, n_redundant=0, random_state=0, shuffle=False)
```

```python
scores = []
n_estimators = np.arange(50,1050,50)
```

```python
for i in n_estimators:
    clf = RandomForestClassifier(max_depth=2, random_state=0, n_estimators=i)
    clf.fit(X, y)
    scores.append(clf.score(X,y))
```

```python
plt.plot(n_estimators, scores)
plt.xlabel('Количество Деревьев')
plt.ylabel('Точность')
plt.title('Метрика качества от количества деревьев')
plt.show()
```

2. Зависимость качества обучения от глубины деревьев

```python
accuracies = []
n_depths = np.arange(2,1000,2)
```

```python
for i in n_depths:
    clf = RandomForestClassifier(max_depth=i, random_state=0)
    clf.fit(X, y)
    accuracies.append(clf.score(X,y))
```

```python
plt.plot(n_depths, accuracies)
plt.xlabel('Глубина')
plt.ylabel('Точность')
plt.title('Качество обучения от глубины деревьев')
plt.show()
```
